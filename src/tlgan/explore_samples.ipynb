```python
print(f['images'].shape)

images_to_plot = 30
fig = plt.figure(figsize=(16, 50))
cols = 2
rows = images_to_plot/cols

for i, image in enumerate(f['images']):
    if i >= images_to_plot: break
    fig.add_subplot(rows, cols, i+1)
    plt.imshow(image)
plt.show()
```
